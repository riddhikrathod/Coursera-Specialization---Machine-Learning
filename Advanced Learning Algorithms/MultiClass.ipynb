### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
np.set_printoptions(precision=2)
import logging
logging.getLogger("tensorflow").setLevel(logging.ERROR)
tf.autograph.set_verbosity(0)
```

### Multiclass Classification
- NNs are often used for classifying data such as:
    - Objects in a photo (dog, cat, etc.)
    - Sentence elements to a part of speech (noun, verb, etc.)
- A NN of this type will have multiple units in its final layer, one for each category
- When an input is applied to the network, the output with the highest value is the category predicted
- If the output is applied to a softmax function, the output of the softmax will provide probs of the input being each category

```python
# make 4-class dataset for classification
classes = 4
m = 100
centers = [[-5, 2], [-2, -2], [1, 2], [5, -2]]
std = 1.0
X_train, y_train = make_blobs(n_samples=m, centers=centers, cluster_std=std,random_state=30)
```

```python
# show classes in data set
print(f"unique classes {np.unique(y_train)}")
# show how classes are represented
print(f"class representation {y_train[:10]}")
# show shapes of our dataset
print(f"shape of X_train: {X_train.shape}, shape of y_train: {y_train.shape}")
```

### Model
- Use a 2 layer NN:
    - Layer 1 = 2 units of Relu
    - Layer 2 = 4 units of Linear
    - Apply softmax and get probs
- Unlike binary classification, the network has 4 outputs, one for each class. The output with the highest value is the predicted class.
- It is more numerically stable to pass the linear outputs to the loss function during training
- Softmax is applied when the model is used for predicting probabilitie


```python
tf.random.set_seed(1234)  # applied to achieve consistent results
model = Sequential(
    [
        Dense(2, activation = 'relu',   name = "L1"),
        Dense(4, activation = 'linear', name = "L2")
    ]
)
```

```python
model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=tf.keras.optimizers.Adam(0.01),
)

model.fit(
    X_train,y_train,
    epochs=200
)
```

```python
# gather the trained parameters from the first layer
l1 = model.get_layer("L1")
W1,b1 = l1.get_weights()
```

```python
# gather the trained parameters from the output layer
l2 = model.get_layer("L2")
W2, b2 = l2.get_weights()
```
